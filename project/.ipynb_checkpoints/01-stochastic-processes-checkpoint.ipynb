```python
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 14

import seaborn as sns
palette = sns.color_palette('tab10')
```

# Wiener process

The Brownian motion (physical phenomenon) is described by the Wiener process $B(t)$ (model), which is a continuous-time stochastic process characterized by Gaussian-distributed independent random increments:

- $B(t+h) - B(t) := \Delta B(t) \sim \mathcal{N}(0, h)$

- $\Delta B(t)$ and $\Delta B(t') = B(t'+h') - B(t')$ are indipendent (where $t < t+h \le t' < t'+h'$)

The following algorithm describes how to simulate a Wiener process:

- $\Delta t := t_{i+1} - t_i$
- $B(t_0) = 0$
- $B(t_{i+1}) = B(t_i) + \Delta B(t_i)$, where $\Delta B(t_i) \sim \mathcal{N}(0, \Delta t)$ and therefore $[\Delta B] = T^{1/2}$

The last step can be written equivalently as:

- $B(t_{i+1}) = B(t_i) + \sqrt{\Delta t} \cdot Z(t_i)$, where $Z(t_i) \sim \mathcal{N}(0, 1)$

In fact, using the change of variable $\Delta B = \sqrt{h} \cdot Z$, it can be shown that:

- $p(\Delta B) \cdot d\Delta B = p(Z) \cdot dZ$

```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)
B = np.zeros(len(t))

# create an instance of a Wiener process
for i in range(1, len(t)):
    # normal-distributed independent random increments
    Z = np.random.normal(0, 1)
    B[i] = B[i-1] + np.sqrt(dt)*Z

plt.plot(t, B, color = palette[0])
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$B_i$ $[T^{1/2}]$')
plt.grid()
plt.show()
```

```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)

n = 1000
B = np.zeros((n, len(t)))

# simulate n Wiener processes
for j in range(0, n):
    # create an instance of a Wiener process
    for i in range(1, len(t)):
        # normal-distributed independent random increments
        Z = np.random.normal(0, 1)
        B[j,i] = B[j,i-1] + np.sqrt(dt)*Z

# calculate the root mean square
rms = np.sqrt((B**2).mean(axis = 0))

plt.plot(t, rms, color = palette[0], label = 'simulation')
plt.plot(t, np.sqrt(t), '--', color = palette[1], label = 'expectation')
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$RMS$ $[T^{1/2}]$')
plt.legend()
plt.grid()
plt.show()
```

# Brownian motion

It is an example of a diffusion process whose path can be described by a stochastic differential equation:

- $dX(t) = \sqrt{2D} \cdot dB(t)$

where $D$ is the diffusion coefficient and $B(t)$ is a Wiener process.

The following algorithm describes how to simulate a Brownian motion:

- $\Delta t := t_{i+1} - t_i$
- $X(t_0) = 0$
- $X(t_{i+1}) = X(t_i) + \sqrt{2D} \cdot \Delta B(t_i)$, where $\Delta B(t_i) \sim \mathcal{N}(0, \Delta t)$

The last step can be written equivalently as:

- $X(t_{i+1}) = X(t_i) + \sqrt{2D} \cdot \sqrt{\Delta t} \cdot Z(t_i)$, where $Z(t_i) \sim \mathcal{N}(0, 1)$

Note that since $[X] = L$ and $[\Delta B] = T^{1/2}$, the physical dimension of the diffusion coefficient must be $[D] = L^2/T$ which is indeed the case.


```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)

D = 10
X = np.zeros(len(t))

# create an instance of a Brownian motion
for i in range(1, len(t)):
    # normal-distributed independent random increments
    Z = np.random.normal(0, 1)
    X[i] = X[i-1] + np.sqrt(2*D*dt)*Z

plt.plot(t, X, color = palette[0])
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$X_i$ $[L]$')
plt.grid()
plt.show()
```

```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)

n = 1000
D = 10
X = np.zeros((n, len(t)))

# simulate n Brownian motions
for j in range(0, n):
    # create an instance of a Brownian motion
    for i in range(1, len(t)):
        # normal-distributed independent random increments
        Z = np.random.normal(0, 1)
        X[j,i] = X[j,i-1] + np.sqrt(2*D*dt)*Z

# calculate the root mean square
rms = np.sqrt((X**2).mean(axis = 0))

plt.plot(t, rms, color = palette[0], label = 'simulation')
plt.plot(t, np.sqrt(2*D*t), '--', color = palette[1], label = 'expectation')
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$RMS$ $[L]$')
plt.legend()
plt.grid()
plt.show()
```

# OrnsteinUhlenbeck process

It is another example of a diffusion process whose path can be described by a stochastic differential equation:

- $dX(t) = f(X(t)) \cdot dt + \sqrt{2D} \cdot dB(t)$

where $f(X(t))$ is a function of $X(t)$ and $B(t)$ is a Wiener process.


From  a physical point of view, it is a stochastic process that correspond to a particle moving in a harmonical potential $U_{ext}(X)$ and subject to thermal noise caused by the medium.

The process can be considered to be a modification of the Brownian motion in which the properties of the process have been changed so that there is a tendency of the path to move back towards a central location, with a greater attraction when the process is further away from the center.

It can be show that $f(X(t))$ corresponds to the external force $F_{ext}$ divided by the friction coefficient $[\gamma] = M/T$. Therefore it is possible to get the form of $f(X(t))$:

- $U_{ext}(X) = \frac{1}{2}m w^2 X^2$
- $F_{ext} = - d_X U_{ext}(X) = - m w^2 X := - k X$
where $[k] = M/T^2$ can be thought as the spring constant

- $f(X(t)) = - \frac{k}{\gamma} \cdot X(t)$
where $[\frac{k}{\gamma}] = T^{-1}$.

The stochastic differential equation becomes:

- $dX(t) = - \frac{k}{\gamma} \cdot X(t) \cdot dt + \sqrt{2D} \cdot dB(t)$

In general the spring rest length $X(0)$ can be different from $0$, thus:

- $dX(t) = \frac{k}{\gamma} \cdot (X_0 - X(t)) \cdot dt + \sqrt{2D} \cdot dB(t)$


For a generic model indipendent of the physical point of view, the stochastic differential equation can be parametrized as follows:

- $dX(t) = \theta \cdot (X_0 - X(t)) \cdot dt + \sigma \cdot dB(t)$

The following algorithm describes how to simulate such OrnsteinUhlenbeck process:

- $\Delta t := t_{i+1} - t_i$
- $X(t_0) = X(t_0=0)$
- $X(t_{i+1}) = X(t_i) + \theta \cdot (X_0 - X(t_i)) \cdot \Delta t + \sigma \cdot \Delta B(t_i)$
where $\Delta B(t_i) \sim \mathcal{N}(0, \Delta t)$.

The last step can be written equivalently as:

- $X(t_{i+1}) = X(t_i) + \theta \cdot (X_0 - X(t_i)) \cdot \Delta t + \sigma \cdot \sqrt{\Delta t} \cdot Z(t_i)$, where $Z(t_i) \sim \mathcal{N}(0, 1)$

Note that if $[X] = L$ and given that $[\Delta B] = T^{1/2}$, the physical dimensions of the parameters are $[\theta] = T^{-1}$ and $[\sigma] = L/T^{1/2}$.

To make the notation consistent with that of Albert:

- $\theta := \tau^{-1}$ where $[\tau] = T$
- $\sigma = \sigma \cdot \sqrt{\tau}/\sqrt{\tau} := \sigma_X/\sqrt{\tau}$ where $[\sigma_X] = L$

Finally:

- $X(t_{i+1}) = X(t_i) + \tau^{-1} \cdot (X_0 - X(t_i)) \cdot \Delta t +  \tau^{-1/2} \cdot \sigma_X \cdot \sqrt{\Delta t} \cdot Z(t_i)$, where $Z(t_i) \sim \mathcal{N}(0, 1)$.

```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)

tau = 100
sigma_X = 10
X0 = 0

X = np.zeros(len(t))
X[0] = 100

# create an instance of a OrnsteinUhlenbeck process
for i in range(1, len(t)):
    # normal-distributed independent random increments
    Z = np.random.normal(0, 1)
    X[i] = X[i-1] + (X0 - X[i-1])*(dt/tau) + sigma_X*np.sqrt(dt/tau)*Z

plt.plot(t, X, color = palette[0])
plt.axhline(y = X0, color = palette[1], linestyle = '--')
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$X_i$ $[L]$')
plt.grid()
plt.show()
```

# Autoregressive model AR(1)

It can be considered as the discrete-time analogue of the OrnsteinUhlenbeck process. Adopting Albert notation:

- $\xi_{i+1} = a + b \cdot \xi_{i} + c \cdot \epsilon_i$

where $-1 < b < 1$ and $\epsilon_i \sim \mathcal{N}(0, 1)$

The relationships between these variables and those introduced for the OrnsteinUhlenbeck process are as follows:

- $\xi_{i} = X(t_i)$
- $\epsilon_i = Z(t_i)$

- $a = \tau^{-1} \cdot X_0 \cdot \Delta t$
- $b = 1 - \tau^{-1} \cdot \Delta t$
- $c = \tau^{-1/2} \cdot \sigma_X \cdot \sqrt{\Delta t}$

Note that he constraint on $b$ implies a relation on $\tau$ and $\Delta t$ that must verify the following:

- $\tau > 0.5 \cdot \Delta t$

```python
# set seed for random number generator
np.random.seed(12345)

dt = 10
t = np.arange(0, 100*dt, dt)

tau = 100
sigma_X = 10
X0 = 0

a = X0*(dt/tau)
b = 1 - (dt/tau)
c = sigma_X*np.sqrt(dt/tau)
print('a:', a, '\nb:', b, '\nc:', c)

X = np.zeros(len(t))
X[0] = 0

# create an instance of a AR(1)
for i in range(1, len(t)):
    # normal-distributed independent random increments
    Z = np.random.normal(0, 1)
    X[i] = a + b*X[i-1] + c*Z

plt.plot(t, X)
plt.axhline(y = X0, color = 'tab:orange', linestyle = '--')
plt.xlabel(r'$t_i$ $[T]$')
plt.ylabel(r'$X_i$ $[L]$')
plt.grid()
plt.show()
```
