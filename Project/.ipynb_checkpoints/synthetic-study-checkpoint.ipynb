# White noise

The most common representation of white noise is a series of independent and identically distributed random samples drawn from a distribution with zero mean and finite variance. If the underlying distribution is normal, the series is called Gaussian white noise.

```python
# set seed for random number generator
np.random.seed(12345)

# generate Gaussian white noise series
noise = pd.Series(np.random.normal(0, 1, 1000))

# summary stats
noise.describe()
```

```python
# line plot
fig, ax = plt.subplots()
ax.plot(noise.index, noise.values, color = palette[0])
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
plt.grid()
plt.show()
```

```python
# histogram
fig, ax = plt.subplots()
ax.hist(noise.values, bins = 'scott', range = (-4, 4), color = palette[0])
ax.set_xlabel('x(t)')
ax.set_ylabel('counts')
plt.grid()
plt.show()
```

### Autocorrelation function

https://www.influxdata.com/blog/autocorrelation-in-time-series-data/

```python
# autocorrelation function
def acf(series, nlag):
    
    # data
    x = np.array([i for i in range(0, nlag+1)])
    y = np.array([series.autocorr(lag=i) for i in range(0, nlag+1)])
    
    # 95% confidence interval
    # note: it can been shown that the autocorrelation coefficient r(k) follows a Gaussian distribution
    # with variance Var(r(k)) = 1/N for all k (first order approximation)
    var = 1/len(series.values)
    ci = 1.96*np.sqrt(var)
    
    # plot
    fig, ax = plt.subplots()
    ax.plot(x, y, '-o', color = palette[0])
    plt.axhline(y = ci, color = 'gray', linestyle = '--', label = '95% confidence interval')
    plt.axhline(y = -ci, color = 'gray', linestyle = '--')
    ax.set_xticks(np.arange(0, nlag+1, step=5))
    ax.set_ylim([-1, 1])
    ax.set_xlabel('lag')
    ax.set_ylabel('autocorrelation')
    plt.legend()
    plt.grid()
    plt.show()
```

# Sinusoidal signal with Gaussian white noise

```python
# set seed for random number generator
np.random.seed(12345)

sampling_rate = 100

# signal frequency
# note: if the highest signal frequency is smaller than the Nyquist frequency (which is half the sampling rate)
# the resulting discrete-time series is said to be free of the distortion 
frequency = sampling_rate/100

# time interval
dt = 1/sampling_rate
t = np.arange(0, 3*(1/frequency)+dt, dt)

# sinusoidal signal
signal = 10*np.sin(2*np.pi*frequency*t)

# Gaussian white noise
noise = 1*np.random.normal(0, 1, len(t))

# sinusoidal signal with Gaussian white noise
x = signal + noise

df = pd.DataFrame(data = {'t':t, 'x(t)':x})
df
```

```python
# line plot
fig, ax = plt.subplots()
ax.plot(df['t'], df['x(t)'], color = palette[0])
ax.set_xlabel('t [s]')
ax.set_ylabel('x(t)')
plt.grid()
plt.show()
```

# Correlated error model 

https://people.duke.edu/~rnau/411arim3.htm

### Moving average model

$x_t = c + \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \dots + \theta_q \cdot \epsilon_{t-p}$

where $\epsilon_t$ is white noise

https://otexts.com/fpp2/MA.html

```python
# simulate correlated error with an moving average model of order q
def movingaverage(n, q = 1, theta = [0.6]):
    '''
    order 'q'
    parameters 'phi':
        if q = 1:
            -1 < theta[0] < 1
        if q = 2:
            -1 < theta[1] < 1
            theta[1] + theta[0] > -1, theta[0] - theta[1] < 1
    '''

    # set seed for random number generator
    np.random.seed(12345)

    # generate Gaussian white noise
    noise = np.random.normal(0, 1, n)

    # initialization
    corr_err = np.zeros(n)
    corr_err[0:q-1] = noise[0:q-1]

    for i in range(q, n):
        corr_err[i] = noise[i]
        for j in range(0, q):
            corr_err[i] = corr_err[i] + theta[j]*noise[i-(j+1)]

    # zero mean
    corr_err = corr_err - np.mean(corr_err)

    return(corr_err)
```

```python
p = 2
phi = [0.6/p for i in range(0, p)]
corr_err = pd.Series(movingaverage(1000, p, phi))

# line plot
fig, ax = plt.subplots()
ax.plot(corr_err.index, corr_err.values, color = palette[0])
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
plt.grid()
plt.show()

# autocorrelation
acf(corr_err, 20)
```

### Autoregressive model

$x_t = c + \phi_1 \cdot x_{t-1} + \dots + \phi_p \cdot x_{t-p} + \epsilon_t$

where $\epsilon_t$ is white noise

https://otexts.com/fpp2/AR.html

https://webspace.maths.qmul.ac.uk/b.bogacka/TimeSeries/TS_Chapter4_5.pdf

```python
# simulate correlated error with an autoregressive model of order p
def autoregressive(n, p = 1, phi = [0.6]):
    '''
    order 'p'
    parameters 'phi':
        if p = 1:
            -1 < phi[0] < 1
        if p = 2:
            -1 < phi[1] < 1
            phi[0] + phi[1] < 1, phi[1] - phi[0] < 1
    '''

    # set seed for random number generator
    np.random.seed(12345)

    # generate Gaussian white noise
    noise = np.random.normal(0, 1, n)

    # initialization
    corr_err = np.zeros(n)
    corr_err[0:p-1] = noise[0:p-1]

    for i in range(p, n):
        corr_err[i] = noise[i]
        for j in range(0, p):
            corr_err[i] = corr_err[i] + phi[j]*corr_err[i-(j+1)]

    # zero mean
    corr_err = corr_err - np.mean(corr_err)

    return(corr_err)
```

```python
p = 2
phi = [0.6/p for i in range(0, p)]
corr_err = pd.Series(autoregressive(1000, p, phi))

# line plot
fig, ax = plt.subplots()
ax.plot(corr_err.index, corr_err.values, color = palette[0])
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
plt.grid()
plt.show()

# autocorrelation
acf(corr_err, 20)
```
